```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df = pd.read_csv('/content/transactions.csv')
```

```python
df.head(5)
```

```python
transactions = df['item_names'].apply(lambda x: x.split(',')).tolist()
```

clean

```python
transactions = df['item_names'].apply(
    lambda x: [item.strip() for item in x.split(',')]
).tolist()
```

trasform 0 ,1

```python
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_data, columns=te.columns_)
df_encoded.head()
```

```python
df_encoded = df_encoded.astype(bool)
```

frequent items (suport)

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_items = fpgrowth(df_encoded, min_support=0.001, use_colnames=True)

print("frequent items =", len(frequent_items))
frequent_items.head(20)
```

*rules*

List to detect relation

```python
rules = association_rules(
frequent_items, metric="lift", min_threshold=1)

print("\nAssociation Rules (Top 20):")
print("rules =", len(rules))

rules.head(20)
```

```python
rules.iloc[: , 0:7].sort_values(by='lift', ascending=False).head(20)
```

```python
def get_closed_patterns(frequent_items):
    closed_patterns = []

    for i, row_i in frequent_items.iterrows():
        is_closed = True

        for j, row_j in frequent_items.iterrows():
            if row_i['itemsets'] < row_j['itemsets']:  # proper subset
                if row_i['support'] <= row_j['support']:
                    is_closed = False
                    break

        if is_closed:
            closed_patterns.append(row_i)

    return pd.DataFrame(closed_patterns)

closed_itemsets = get_closed_patterns(frequent_items)
print("Closed Itemsets (Top 20):")
closed_itemsets.head(20)
```

```python
def get_max_patterns(frequent_items):
    max_patterns = []

    for i, row_i in frequent_items.iterrows():
        is_max = True

        for j, row_j in frequent_items.iterrows():
            if row_i['itemsets'] < row_j['itemsets']:  # proper subset
                is_max = False
                break

        if is_max:
            max_patterns.append(row_i)

    return pd.DataFrame(max_patterns)
max_itemsets = get_max_patterns(frequent_items)
print("Maximal Itemsets (Top 20):")
max_itemsets.head(20)
```
